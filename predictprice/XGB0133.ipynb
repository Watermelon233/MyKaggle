```python
import pandas as pd
import numpy as np
train = pd.read_csv('train.csv',index_col='Id')
test = pd.read_csv('test.csv',index_col='Id')
print(train.shape,test.shape)
```

```python
data = pd.concat([train.drop('SalePrice',1),test])
```

```python
target = train['SalePrice'].copy()
```

```python
nummiss = data.isnull().sum()
```

```python
data = data.drop(nummiss[nummiss>500].index,1)
```

```python
data.info()
```

```python
print(data['MSZoning'].dtype)
```

```python
itemlist={}
for col in data:
    if data[col].dtype == object:
        itemlist[col] = list(set(data[col]))

```

```python
itemlist
```

```python
isConti={}
isConti[1] = ['BsmtCond','BsmtExposure', 'BsmtFinType1','BsmtFinType2', 'BsmtQual','Electrical','ExterCond', 'ExterQual','Functional','GarageCond','GarageFinish','GarageQual','HeatingQC','KitchenQual','LandContour']
isConti[0] = ['BldgType','CentralAir','Condition1','Condition2', 'Exterior1st', 'Exterior2nd','Foundation','GarageType','Heating','HouseStyle','LandSlope','LotConfig','LotShape', 'MSZoning','MasVnrType','Neighborhood','PavedDrive','RoofMatl','RoofStyle','SaleType','Street','Utilities','SaleCondition']
```

```python
disVar = data[isConti[0]].copy()
disVar = pd.get_dummies(disVar)
disVar.head()
```

```python
enumConti = []
for key in isConti[1]:
    enumConti.extend(itemlist[key])
enumConti = set(enumConti)
enumConti
enumNum = {'nan':np.NaN,'Bnk':2,'Mn':1,'Typ':7,'FuseF':2,'Lvl':3,'Unf':0,'Min1':6,
 'Sev':1,
 'Low':0,
 'Mix':0,
 'RFn':1,
 'Gd':3,
 'TA':2,
 'LwQ':1,
 'Ex':4,
 'Maj2':2,
 'ALQ':4,
 'Po':0,
 'No':0,
 'Mod':4,
 'FuseP':1,
 'HLS':1,
 'BLQ':3,
 'FuseA':3,
 'Av':2,
 'Min2':5,
 'Rec':2,
 'SBrkr':4,
 'Fa':1,
 'Maj1':3,
 'Fin':2,
 'GLQ':5}
```

```python
ContiVar = data[isConti[1]].copy()
ContiVar = ContiVar.replace(enumNum)
```

```python
nocatcol = []
for col in data:
    if data[col].dtype != object:
        nocatcol.append(col)
Numericdata = pd.concat([data[nocatcol].copy(),ContiVar,disVar],axis=1)
Numericdata
```

```python
from sklearn.preprocessing import Imputer
my_imputer = Imputer()
#transformå’Œfit_transformçš„ç»“æžœæ˜¯å®Œå…¨ä¸€æ ·çš„ï¼Œä½†æ˜¯éœ€è¦å…ˆfitå¾—åˆ°æ¨¡åž‹çš„å„ç§å‚æ•°ï¼Œå› æ­¤å…ˆåœ¨è®­ç»ƒé›†è°ƒç”¨fit_transform
imputed_X_train = my_imputer.fit_transform(Numericdata[:1460])
imputed_X_test = my_imputer.transform(Numericdata[1460:])
y_train = target
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(imputed_X_train,y_train, test_size=0.25)
```

```python
from xgboost import XGBRegressor
my_model = XGBRegressor()
my_model.fit(train_X, train_y, verbose=False)
```

```python
Pval_y = my_model.predict(val_X)
```

```python
mean_absolute_error(Pval_y,val_y)
```

```python
my_model.score(val_X,val_y)
```

```python
y_test = my_model.predict(imputed_X_test)
y_test.shape
```

```python
subdata = pd.DataFrame({'Id':test.index,'SalePrice':y_test})
```

```python
subdata.to_csv('submission.csv', index=False)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_model, train_X, train_y, scoring='neg_mean_absolute_error')
print(scores)
```
